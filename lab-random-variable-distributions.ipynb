```python
#Importing the libraries
import pandas as pd
import numpy as np
import datetime
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\Samu\Documents\Ironhack\lab-random-variable-distributions\files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
#Getting the numerical variables
numerical = df.select_dtypes(include = np.number)
```

```python
#Checking if the variables are normally distributed using a distribution plot
for col in numerical.columns:
    sns.distplot(numerical[col])
    plt.show()
```

Customer lifetime value and months since policy inception are the only variables that are normally distributed. 

```python
#Applying the logarithm to the variables to see if they are normally distributed
for col in numerical.columns:
    numerical[col] = np.log(numerical[col]+1)
    sns.distplot(numerical[col])
    plt.show()
```

Applying log transformation completely messes the distributions, making some of them normally distributed byt really skewed, and the original normally distributed ones ceases to be so.

```python
#Applying square root to the variables to see if they are normally distributed
for col in numerical.columns:
    numerical[col] = np.sqrt(numerical[col])
    sns.distplot(numerical[col])
    plt.show()
```

With square root transformation the variables are more even, but narrower and customer lifetime value also ceases to be normally distributed.
